```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
metrics = []
with open(r'./metrics2.txt', 'r') as fp:
    for line in fp:
        x = line[:-1]
        metrics.append(x)
```

```python
loss_train = np.array(metrics[0:51]).astype(np.float32)
loss_val = np.array(metrics[51:102]).astype(np.float32)

auc_train = np.array(metrics[102:153]).astype(np.float32)
auc_val = np.array(metrics[153:204]).astype(np.float32)

acc_train = np.array(metrics[204:255]).astype(np.float32)
acc_val = np.array(metrics[255:306]).astype(np.float32)

x_axis = list(range(0,510,10))
```

```python
fig, (ax1) = plt.subplots(1, 1, figsize=(10,10))
ax1.plot(x_axis, loss_train, label="Training Loss")
ax1.plot(x_axis, loss_val, label="Validation Loss")
ax1.legend()

ax1.plot(x_axis, auc_train, label="Training AUC")
ax1.plot(x_axis, auc_val, label="Validation AUC")

ax1.plot(x_axis, acc_train, label="Training Accuracy")
ax1.plot(x_axis, acc_val, label="Validation Accuracy")
ax1.axvline(x=330, label="Early Stopping")
ax1.legend()

plt.show()
```
